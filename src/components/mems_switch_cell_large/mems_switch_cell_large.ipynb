# MEMS Switch Cell — 1.75x Scale (75% Larger)

Same design as `mems_switch_cell` but with macro dimensions scaled by **1.75x**.

**Scaled (x1.75):** proof mass dimensions, spring length/gap, finger length,
number of finger pairs, holder heights, anchor widths, pad sizes,
inter-comb spacing, layout insets, finger overlap distance.

**Fixed:** hole diameter (5 um), hole gap (0.8 um), finger width (0.5 um),
finger gap (0.5 um), spring beam width (0.5 um), structural linewidth (0.8 um),
anchor hole size (2 um), layer overhangs (0.3/0.5 um).

| Variant | 1x Size | 1.75x Size | Ratio |
|---------|---------|------------|-------|
| Multi comb (width) | 112 um | 194 um | 1.72x |
| Multi comb (height) | 130 um | 229 um | 1.76x |

```python
import sys
sys.path.insert(0, "../../libraries")

from mcw_custom_optical_mems_pdk import PDK
PDK.activate()

from mems_switch_cell_large import (
    make_switch_mems_large,
    make_switch_mems_multi_large,
    make_switch_mems_cut_large,
    S,
)
```

## Single Comb Pair — 1.75x

```python
sw1 = make_switch_mems_large()
bb = sw1.dbbox()
print(f"Cell size: {bb.width():.1f} x {bb.height():.1f} um")
sw1.plot()
```

## Multi Comb Groups — 1.75x

```python
sw3 = make_switch_mems_multi_large()
bb = sw3.dbbox()
print(f"Cell size: {bb.width():.1f} x {bb.height():.1f} um")
sw3.plot()
```

## Multi Comb Groups + Corner-Cut Anchors — 1.75x

```python
sw3c = make_switch_mems_cut_large(
    shape_code_array=[0, 0, 1],
    holder_fixed_leftcut_array=[0, 0, 70],
    holder_fixed_topcut_array=[0, 0, 17.5],
)
bb = sw3c.dbbox()
print(f"Cell size: {bb.width():.1f} x {bb.height():.1f} um")
sw3c.plot()
```

## Side-by-Side Comparison

```python
sys.path.insert(0, "../mems_switch_cell")
from mems_switch_cell import make_switch_mems_multi

sw1x = make_switch_mems_multi(num_combs=3, comb_edge_lower_array=[4, 34, 64])
bb1x = sw1x.dbbox()
bb175 = sw3.dbbox()

print(f"1x cell:    {bb1x.width():.1f} x {bb1x.height():.1f} um")
print(f"1.75x cell: {bb175.width():.1f} x {bb175.height():.1f} um")
print(f"Ratio:      {bb175.width()/bb1x.width():.2f}x width, {bb175.height()/bb1x.height():.2f}x height")
```
