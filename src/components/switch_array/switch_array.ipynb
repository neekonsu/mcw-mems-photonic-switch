# 8x8 MEMS Switch Array

64 single-comb-pair 1.75x switch cells arranged at **250 um pitch** with routed POLY_MEMS interconnects.

**Routing scheme:**
- **Per-cell stubs** (20 um): each anchor gets a short trace extending **away from cell center**
  - Upper spring anchor → stub going UP
  - Lower spring anchor → stub going DOWN
  - Right comb anchor → stub going RIGHT
  - Left comb anchor → stub going LEFT
- **Stiction-prevention anchors**: 10x10 um, 3-layer (POLY_TOP + POLY_MEMS + SI_FULL), one at each stub endpoint
- **No overlap**: stubs clear of springs, shuttles, and comb drives
